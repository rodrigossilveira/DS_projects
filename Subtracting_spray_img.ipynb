```python
import math
import numpy as np
import matplotlib.pyplot as plt

threshold = 7
#Getting inputs the code need for the calculations
R11 = []
#R11.append(float(input('Distância inicial do orifício do eletroinjetor(mm) ')))
#R111 = float(input('Incrementos de distância a partir dadistância inicial(mm) '))
#R1111 = float(input('Distância final '))
R11.append(5)
R111 = 0.1
R1111 = 20.0

angulo1, angulo2 = 0,0
angulo, erro, angulomaiserro, angulomenoserro, angulo3 = [], [], [], [], []
ang1, ang2 = [], []
angulo4 = []
#angulo4 = [None]*int((R1111-R11[0])/R111)

# Transforming the distance from mm to pixels
R1 = R11[0]*13.907
print("R1 = {a}".format(a=R1))

i,j,k = 29,393,0
dx = [None]
dx[0] = 0
e=0

#while R1<R1111*13.907:
for step in np.arange(R1,R1111*13.907,R111*13.907):
  a, b, c, d = 0, 0, 1, 0
  k = 0
  X1 = []
  dx = []
  dx=[0]
  
  
  #Scanning the image
  for theta in np.arange(29*math.pi/36, math.pi/4, -math.pi/360):
    #a=0
    k+=1
    angulo3.append(theta*(180/math.pi))
    x,y = round(i+step*math.sin(theta)), round(j+step*math.cos(theta))    
    X1.append(float(img_normalized2[x][y]))
    # xreal(k)=double(x1(k));
    #E(round(i+R1*sin(teta)),round(j+R1*cos(teta)))
    #Spray_frontier[x][y] = 255

    #Determining the gradient
    if k>=2 :
      dx.append(X1[k-1] - X1[k -2])
      #dx.append((X1[k-1] - X1[k -2])/((math.pi/360)*R1))
    #If the gradient is bigger than a threshold save the angle the cone starts
    if abs(dx[k-1])>=threshold and a == 0:
      angulo1, a = theta, 1
      ang1.append(angulo1)
      Spray_frontier[x][y] = 255

      #break
  for l in range(k,1,-1):
    teta2 = 29*math.pi/36-l*(math.pi/360)
    if abs(dx[l-1])>=threshold and b==0:
      angulo2,b = teta2, 1
      ang2.append(angulo2)
      x,y = round(i+step*math.sin(teta2)), round(j+step*math.cos(teta2)) 
      Spray_frontier[x][y] = 255
  d+=1
  angulo4.append((angulo1-angulo2)*(180/math.pi))
  #R1=R1+R111*13.907
  e+=1 
    #if e >=1 :
  R11.append(R11[e-1]+R111)
  angulo.append(np.mean(angulo4))
  erro.append(2.045*(np.std(angulo4))/(30**0.5))
  angulomaiserro.append(angulo[e-1]+erro[e-1])
  angulomenoserro.append(angulo[e-1]-erro[e-1])

plt.plot(R11[:-1],angulo4,'o')    
#plt.plot(R11[:-1],angulo,'o',R11[:-1],angulomaiserro,'x',R11[:-1],angulomenoserro,'+',)
plt.grid(True)
plt.ylabel('Ângulo de cone medido')
plt.xlabel('Distânca do spray')
plt.ylim(0,80)
plt.show()

plt.imshow(Spray_frontier, interpolation= 'none')
#plt.ylim(66.49,66.51)1
#plt.errorbar(R11[:-1],angulo, xerr = erro, yerr = None, fmt='o', ecolor = 'red',color='yellow')
```
